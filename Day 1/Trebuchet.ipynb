```python
import re

class Trebuchet:
    def __init__(self, filename: str):
        self.filename = filename
        self.mappings = {
            
            "one": 1,
            "two": 2,
            "three": 3,
            "four": 4,
            "five": 5,
            "six": 6,
            "seven": 7,
            "eight": 8,
            "nine": 9,
        }
    
        
    def part1(self):
        x: list = []
        with open(self.filename, "r") as file:
            while True:
                line = file.readline()
                if not line:
                    break
                x.append(line[:-1])
                
        sum: int = 0
        
        for i in range(0, len(x)):
            first: int = 0
            last: int = 0
            for char in x[i]:
                if char.isnumeric() and first == 0:
                    first = int(char)
                    last = int(char)
                elif char.isnumeric():
                    last = int(char)
            sum += int(str(first) + str(last))
            
        print("Part 1 answer is: ", sum)
        
    def part2(self):
        x: list = []
        with open(self.filename, "r") as file:
            while True:
                line = file.readline()
                if not line:
                    break
                x.append(line[:-1])
        
        sum: int = 0
        for i, string in enumerate(x):
            digitlist = []
            for j, char in enumerate(string):
                if char.isnumeric():
                    digitlist.append((char, j))
            
            for key in self.mappings.keys():
                for indices in re.finditer(key, string):
                    startSpan:int = indices.span()[0]
                    digitlist.append((str(self.mappings[key]), startSpan))
                    
            digitlist = sorted(digitlist, key=lambda x: x[1])
            print(str(i+1) + " " + digitlist[0][0] + digitlist[-1][0])
            sum += int(digitlist[0][0] + digitlist[-1][0])
            
            
        print("Part 2 answer is: ", sum)
        
```

```python
trebuchet = Trebuchet("input_part2.txt")
trebuchet.part2()
```
